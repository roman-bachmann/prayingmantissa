```python
import numpy as np
import seaborn as sb
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Setup

raw = pd.read_csv('./data/interactive_data.csv')
frame = pd.DataFrame()
sb.set()

width, height = 192, 175

deaths = raw.iloc[0]['Deaths']
```

```python
# Helper methods

def empty_grid():
    return np.zeros((height, width))

def draw(title, plot):
    color_palette = sb.color_palette("RdYlBu", 10)
    
    _, ax = plt.subplots(figsize=(20, 20))
    ax.set_title(title, fontsize=25)
    
    sb.heatmap(plot, cbar=False, xticklabels=[], yticklabels=[], ax=ax, linewidths=1.5, cmap=color_palette)
```

```python
# Plots

def plot1():
    """Total number of suicides amongst all gunshot deaths."""
    suicides = raw.loc[108]['Deaths']
    
    grid = empty_grid()
    
    count = 0
    for x in range(width):
        for y in range(height):
            if count < suicides:
                grid[y][x] = 0
            else:
                grid[y][x] = 1
            count += 1
    
    return grid
```

```python
draw('#Suicides amongst gunshot deaths', plot1())
```
